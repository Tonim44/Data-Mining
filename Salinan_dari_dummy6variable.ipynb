```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn.metrics as metrics
import matplotlib
matplotlib.use('Agg')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae
from sklearn import preprocessing
```

```python
# import dan baca data
path_="/content/gdrive/My Drive/Colab Notebooks/dataset6variabel.csv"
dataset=pd.read_csv(path_)
dataset.head()
```

```python
#dummy variabel
dummy_variabel_1 = pd.get_dummies(dataset["Nama Merk"])
dummy_variabel_1.head()
```

```python
datasetdummy = pd.concat([dummy_variabel_1,dataset], axis=1)

datasetdummy.drop("Nama Merk", axis=1, inplace=True)
```

```python
datasetdummy.head()
```

```python
#mendefinisikan nilai x dan y
x = datasetdummy.iloc[:, 0:12].values
x_dataset = pd.DataFrame(x)
x_dataset.head()
```

```python
y = datasetdummy.iloc[:, 12].values
y_dataset = pd.DataFrame(y)
y_dataset.head()
```

```python
#Pembagian Data
X_train,X_test,y_train,y_test, =train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
#Data latih X & Y
headerList = ['D1','D2','D3','D4','D5','D6','X1', 'X2', 'X3', 'X4', 'X5', 'X6']
X_train_dataset = pd.DataFrame(X_train, columns=headerList)
X_train_dataset
```

```python
headerList = ['Y']
y_train_dataset = pd.DataFrame(y_train, columns=headerList)
y_train_dataset
```

```python
#Matriks X
X = pd.DataFrame(X_train_dataset)
X['1'] = 1
X.columns.tolist()
X = X[['1','D1','D2','D3','D4','D5','D6','X1', 'X2', 'X3', 'X4', 'X5', 'X6']]
X
```

```python
#Matriks Transpose X
XT = X.transpose()
XT
```

```python
#Memulai MLR
reg1 = LinearRegression()
reg1.fit(X_train,y_train)
```

```python
reg1.intercept_
```

```python
#reg1.coef_
koefisien = pd.DataFrame(reg1.coef_)
koefisien
```

```python
#Data Uji (Test)
headerList = ['D1','D2','D3','D4','D5','D6','X1', 'X2', 'X3', 'X4', 'X5', 'X6']
Test_dataset = pd.DataFrame(X_test,columns=headerList)
Test_dataset
```

```python
headerList = ['Y']
y_test_dataset = pd.DataFrame(y_test, columns=headerList)
y_test_dataset
```

```python
#Hasil Prediksi Test
y_pred_test = reg1.predict(X_test)
pred_test_dataset = pd.DataFrame({'Harga':y_test, 'Hasil Prediksi':y_pred_test})
pred_test_dataset
```

```python
#mencari R2
Cari_SSres = pd.DataFrame({'Nilai Aktual train(Y)':y_test,'Nilai Prediksi Train(Y")':y_pred_test,'(Y-Y")^2':(y_test-y_pred_test)**2})
Cari_SSres
```

```python
SSres = sum((y_test-y_pred_test)**2)
print ("SSres", SSres)
```

```python
#Mencari Yavg
from numpy.ma.extras import average
Yavg = average(y_test)
print (Yavg)
```

```python
#mencari SStot
Cari_SStot = pd.DataFrame({'Nilai Aktual train(Y)':y_test,'Nilai Prediksi Train(Y")':y_pred_test,'Y-Yavg':(y_test-Yavg)**2})
Cari_SStot
```

```python
SStot = sum((y_test-Yavg)**2)
print ("SStot", SStot)
```

```python
#Evaluasi model R-Square
r2 = 1-(SSres/SStot)
print ("R2", r2)
```

```python
#Mencari R2 Adj
r2_adj = 1-(1-r2)*((19-1)/(19-12-1))
print ("R2 Adj", r2_adj)
```

```python
#mencari nilai selisih Error Test
error_test_dataset = pd.DataFrame({'Nilai Aktual Test':y_test,'Nilai Prediksi Test':y_pred_test,'Error':(abs(y_test-y_pred_test))})
error_test_dataset
```

```python
#Evaluasi model MAE MAPE Test
from sklearn.metrics import mean_absolute_error as mae
mae_mape_test = pd.DataFrame({'Nilai Aktual Test(Y)':y_test,'Nilai Prediksi Test(Y")':y_pred_test,'|Y-Y"|':(abs(y_test-y_pred_test)), '(|Y-Y"|/Y)*100':(abs(y_test-y_pred_test)/y_test)*100})
mae_mape_test
```

```python
mae_test = average(abs(y_test-y_pred_test))
print ("MAE", mae_test)
```

```python
mape_test = average(((abs(y_test-y_pred_test))/y_test)*100)
print("MAPE:",mape_test)
```
